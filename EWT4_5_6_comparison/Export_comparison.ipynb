```python
# Exponentieller Fit für Vergleichskurven (ilacco1_0, ilacco1_1, ilacco1_2)
# Fit-Funktion: y = A*exp(-k*t) + C
from scipy.optimize import curve_fit

def exp_decay(t, A, k, C):
    """Exponentielle Zerfallsfunktion: y = A*exp(-k*t) + C"""
    return A * np.exp(-k * t) + C

# Definiere die drei Gruppen mit allen DataFrames
all_data = [
    # ilacco1_0
    [('MCT1', 'o', E_MCT1_ilacco1_0),
     ('BSG_MCT1', 's', E_BSG_MCT1_ilacco1_0),
     ('PfFNT', '^', E_PfFNT_ilacco1_0),
     ('PfFNT_FP', 'D', E_PfFNT_FP_ilacco1_0)],
    
    # ilacco1_1
    [('MCT1', 'o', E_MCT1_ilacco1_1),
     ('BSG_MCT1', 's', E_BSG_MCT1_ilacco1_1),
     ('PfFNT', '^', E_PfFNT_ilacco1_1)],
    
    # ilacco1_2
    [('MCT1', 'o', E_MCT1_ilacco1_2),
     ('BSG_MCT1', 's', E_BSG_MCT1_ilacco1_2),
     ('PfFNT', '^', E_PfFNT_ilacco1_2)]
]

titles = ['ilacco1_0', 'ilacco1_1', 'ilacco1_2']

# Erstelle 1x3 Subplot-Grid für die drei Bedingungen
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Sammle alle Daten für uniforme Skalierung
all_times = []
all_values = []

for idx, (data_group, title) in enumerate(zip(all_data, titles)):
    ax = axes[idx]
    
    for name, marker, df in data_group:
        # Extrahiere Daten
        times = df['Time'].values
        means = df['Mittelwerte'].values
        stds = df['Standardabweichung'].values
        
        # Entferne NaN-Werte
        mask = ~(np.isnan(times) | np.isnan(means))
        times_clean = times[mask]
        means_clean = means[mask]
        stds_clean = stds[mask]
        
        # Sammle für Skalierung
        all_times.extend(times_clean)
        all_values.extend(means_clean)
        
        # Plotte Originaldaten
        ax.errorbar(times_clean, means_clean, yerr=stds_clean,
                    marker=marker, linestyle='', capsize=4,
                    markerfacecolor='white', markeredgewidth=1.5, markersize=6,
                    label=f'{name} (Daten)', alpha=0.7)
        
        # Schätze initiale Parameter für den Fit
        A_guess = means_clean[0] - means_clean[-1] if len(means_clean) > 0 else 1.0
        C_guess = means_clean[-1] if len(means_clean) > 0 else 0.0
        k_guess = 0.01
        
        try:
            # Führe exponentiellen Fit durch
            popt, pcov = curve_fit(exp_decay, times_clean, means_clean,
                                  p0=[A_guess, k_guess, C_guess],
                                  maxfev=10000)
            
            A_fit, k_fit, C_fit = popt
            
            # Erstelle Fit-Kurve
            t_fit = np.linspace(times_clean.min(), times_clean.max(), 200)
            y_fit = exp_decay(t_fit, A_fit, k_fit, C_fit)
            
            # Plotte Fit mit derselben Farbe wie die Daten
            line = ax.plot(t_fit, y_fit, '-', linewidth=2, label=f'{name} (Fit: k={k_fit:.4f})')
            
        except Exception as e:
            print(f'Fit für {name} in {title} fehlgeschlagen: {str(e)}')
    
    # Formatierung
    ax.set_title(f'Efflux Comparison: {title}', fontsize=12)
    ax.set_xlabel('Time [s]')
    ax.set_ylabel(r'$\Delta F/F_0$', rotation=0, labelpad=20)
    ax.yaxis.set_label_coords(-0.15, 0.5)
    ax.grid(alpha=0.3)
    ax.legend(loc='best', fontsize=8, ncol=2)

# Setze uniforme Achsenskalierung für alle Subplots
if all_times and all_values:
    x_min, x_max = min(all_times), max(all_times)
    y_min, y_max = min(all_values), max(all_values)
    x_margin = (x_max - x_min) * 0.05
    y_margin = (y_max - y_min) * 0.1
    
    for ax in axes:
        ax.set_xlim(x_min - x_margin, x_max + x_margin)
        ax.set_ylim(y_min - y_margin, y_max + y_margin)

plt.suptitle('Protein Comparison: Exponential Fits (y = A·exp(-k·t) + C)', fontsize=14, y=1.02)
plt.tight_layout()

# Speichern
os.makedirs('figueres', exist_ok=True)
outpath = os.path.join('figueres', 'Efflux_comparison_with_fits.png')
plt.savefig(outpath, dpi=300, bbox_inches='tight')
print(f'Saved comparison with fits: {outpath}')
plt.show()
```

```python
# Exponentieller Fit für Vergleichskurven mit linearem Fit für PfFNT_FP
from scipy.optimize import curve_fit

def exp_decay(t, A, k, C):
    """Exponentielle Zerfallsfunktion: y = A*exp(-k*t) + C"""
    return A * np.exp(-k * t) + C

# Datengruppe mit Fit-Typ
data_with_fits = [
    ('MCT1', 'o', E_MCT1_ilacco1_0, 'exp'),
    ('BSG_MCT1', 's', E_BSG_MCT1_ilacco1_0, 'exp'),
    ('PfFNT', '^', E_PfFNT_ilacco1_0, 'exp'),
    ('PfFNT_FP', 'D', E_PfFNT_FP_ilacco1_0, 'linear')
]

# Erstelle Plot
fig, ax = plt.subplots(figsize=(14, 7))

# Sammle alle Daten
all_times = []
all_values = []

for name, marker, df, fit_type in data_with_fits:
    # Extrahiere Daten
    times = df['Time'].values
    means = df['Mittelwerte'].values
    stds = df['Standardabweichung'].values
    
    # Entferne NaN-Werte
    mask = ~(np.isnan(times) | np.isnan(means))
    times_clean = times[mask]
    means_clean = means[mask]
    stds_clean = stds[mask]
    
    # Sammle für Skalierung
    all_times.extend(times_clean)
    all_values.extend(means_clean)
    
    # Plotte Originaldaten
    ax.errorbar(times_clean, means_clean, yerr=stds_clean,
                marker=marker, linestyle='', capsize=4,
                markerfacecolor='white', markeredgewidth=1.5, markersize=6,
                label=f'{name} (Daten)', alpha=0.7)
    
    try:
        if fit_type == 'linear':
            # Linearer Fit: y = m*t + b
            coeffs = np.polyfit(times_clean, means_clean, 1)
            m_fit, b_fit = coeffs
            
            # Berechne Unsicherheiten
            y_pred = np.polyval(coeffs, times_clean)
            residuals = means_clean - y_pred
            s_err = np.sqrt(np.sum(residuals**2) / (len(times_clean) - 2))
            t_mean = np.mean(times_clean)
            s_xx = np.sum((times_clean - t_mean)**2)
            m_err = s_err / np.sqrt(s_xx)
            
            # Erstelle Fit-Kurve
            t_fit = np.linspace(times_clean.min(), times_clean.max(), 200)
            y_fit = m_fit * t_fit + b_fit
            
            # Plotte Fit
            ax.plot(t_fit, y_fit, '-', linewidth=2, label=f'{name} (Fit: m={m_fit:.6f})')
        else:
            # Exponentieller Fit
            A_guess = means_clean[0] - means_clean[-1] if len(means_clean) > 0 else 1.0
            C_guess = means_clean[-1] if len(means_clean) > 0 else 0.0
            k_guess = 0.01
            
            popt, pcov = curve_fit(exp_decay, times_clean, means_clean,
                                  p0=[A_guess, k_guess, C_guess],
                                  maxfev=10000)
            
            A_fit, k_fit, C_fit = popt
            
            # Erstelle Fit-Kurve
            t_fit = np.linspace(times_clean.min(), times_clean.max(), 200)
            y_fit = exp_decay(t_fit, A_fit, k_fit, C_fit)
            
            # Plotte Fit
            ax.plot(t_fit, y_fit, '-', linewidth=2, label=f'{name} (Fit: k={k_fit:.4f})')
        
    except Exception as e:
        print(f'Fit für {name} fehlgeschlagen: {str(e)}')

# Formatierung
ax.set_title('Efflux Comparison ilacco1_0: Exponential & Linear Fits', fontsize=14)
ax.set_xlabel('Time [s]', fontsize=12)
ax.set_ylabel(r'$\Delta F/F_0$', rotation=0, labelpad=20, fontsize=12)
ax.yaxis.set_label_coords(-0.08, 0.5)
ax.grid(alpha=0.3)
ax.legend(loc='best', fontsize=9, ncol=2)
plt.tight_layout()

# Speichern
os.makedirs('figueres', exist_ok=True)
outpath = os.path.join('figueres', 'Efflux_ilacco1_0_comparison_with_fits.png')
plt.savefig(outpath, dpi=300, bbox_inches='tight')
print(f'Saved comparison with fits: {outpath}')
plt.show()
```

```python
# Histogramm: Vergleich der Transportraten (k für exponentielle Fits, m für linearen Fit)
from scipy.optimize import curve_fit

def exp_decay(t, A, k, C):
    """Exponentielle Zerfallsfunktion: y = A*exp(-k*t) + C"""
    return A * np.exp(-k * t) + C

# Datengruppe mit Fit-Typ
data_with_fits = [
    ('MCT1', 'o', E_MCT1_ilacco1_0, 'exp'),
    ('BSG_MCT1', 's', E_BSG_MCT1_ilacco1_0, 'exp'),
    ('PfFNT', '^', E_PfFNT_ilacco1_0, 'exp'),
    ('PfFNT_FP', 'D', E_PfFNT_FP_ilacco1_0, 'linear')
]

# Listen für Parameter
names = []
transport_values = []
transport_errors = []
b_values = []
b_errors = []

# Führe Fits durch und sammle Parameter
for name, marker, df, fit_type in data_with_fits:
    times = df["Time"].values
    means = df["Mittelwerte"].values
    
    # Entferne NaN-Werte
    mask = ~(np.isnan(times) | np.isnan(means))
    times_clean = times[mask]
    means_clean = means[mask]
    
    try:
        if fit_type == 'linear':
            # Linearer Fit: y = m*t + b
            coeffs = np.polyfit(times_clean, means_clean, 1)
            m_fit, b_fit = coeffs
            
            # Berechne Unsicherheiten
            y_pred = np.polyval(coeffs, times_clean)
            residuals = means_clean - y_pred
            s_err = np.sqrt(np.sum(residuals**2) / (len(times_clean) - 2))
            t_mean = np.mean(times_clean)
            s_xx = np.sum((times_clean - t_mean)**2)
            m_err = s_err / np.sqrt(s_xx)
            b_err = s_err * np.sqrt(1/len(times_clean) + t_mean**2/s_xx)
            
            # Speichere Werte (Betrag von m als Transportrate)
            names.append(name)
            transport_values.append(abs(m_fit))
            transport_errors.append(m_err)
            b_values.append(b_fit)
            b_errors.append(b_err)
        else:
            # Exponentieller Fit
            A_guess = means_clean[0] - means_clean[-1]
            C_guess = means_clean[-1]
            k_guess = 0.01
            
            popt, pcov = curve_fit(exp_decay, times_clean, means_clean, 
                                  p0=[A_guess, k_guess, C_guess],
                                  maxfev=10000)
            
            A_fit, k_fit, C_fit = popt
            
            # Berechne Unsicherheiten
            k_err = np.sqrt(pcov[1,1])
            C_err = np.sqrt(pcov[2,2])
            
            # Speichere Werte (k als Transportrate, C entspricht b)
            names.append(name)
            transport_values.append(k_fit)
            transport_errors.append(k_err)
            b_values.append(C_fit)
            b_errors.append(C_err)
        
    except Exception as e:
        print(f'Fit für {name} fehlgeschlagen: {e}')

# Erstelle zwei Histogramme nebeneinander
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Histogramm 1: Transportraten (k oder |m|)
x_pos = np.arange(len(names))
bars1 = ax1.bar(x_pos, transport_values, yerr=transport_errors, capsize=5, 
                color='steelblue', edgecolor='black', linewidth=1.5, alpha=0.7)
ax1.set_xlabel('Protein/Bedingung', fontsize=12)
ax1.set_ylabel('Transportrate [1/s]', fontsize=12)
ax1.set_title('Vergleich der Transportraten (k bzw. |m|)', fontsize=14, fontweight='bold')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(names, rotation=45, ha='right', fontsize=10)
ax1.grid(axis='y', alpha=0.3)

# Füge Werte über den Balken hinzu
for i, (bar, val, err) in enumerate(zip(bars1, transport_values, transport_errors)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + err,
            f'{val:.5f}', ha='center', va='bottom', fontsize=8)

# Histogramm 2: Offset-Werte (C oder b)
bars2 = ax2.bar(x_pos, b_values, yerr=b_errors, capsize=5,
                color='coral', edgecolor='black', linewidth=1.5, alpha=0.7)
ax2.set_xlabel('Protein/Bedingung', fontsize=12)
ax2.set_ylabel('Offset (C bzw. b)', fontsize=12)
ax2.set_title('Vergleich der Offset-Werte', fontsize=14, fontweight='bold')
ax2.set_xticks(x_pos)
ax2.set_xticklabels(names, rotation=45, ha='right', fontsize=10)
ax2.grid(axis='y', alpha=0.3)

# Füge Werte über den Balken hinzu
for i, (bar, val, err) in enumerate(zip(bars2, b_values, b_errors)):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + err,
            f'{val:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()

outpath = os.path.join('figueres', 'Efflux_ilacco1_0_parameter_comparison.png')
plt.savefig(outpath, dpi=300, bbox_inches='tight')
print(f'Saved parameter comparison: {outpath}')
plt.show()

# Drucke Zusammenfassung
print('\n=== Fit-Parameter Zusammenfassung (ilacco1_0) ===')
print(f'{"Protein/Bedingung":<25} {"Transportrate":<20} {"Offset":<20}')
print('-' * 70)
for name, t_val, t_err, b_val, b_err in zip(names, transport_values, transport_errors, b_values, b_errors):
    print(f'{name:<25} {t_val:.6f}±{t_err:.6f}   {b_val:.3f}±{b_err:.3f}')
```

```python
# Histogramm 2: Vergleich jeder Probe über verschiedene ilacco-Varianten
from scipy.optimize import curve_fit
import numpy as np

def exp_decay(t, A, k, C):
    """Exponentielle Zerfallsfunktion: y = A*exp(-k*t) + C"""
    return A * np.exp(-k * t) + C

# Definiere alle Proben und ihre Daten über die drei ilacco-Varianten
probe_data = {
    'MCT1': [
        ('ilacco1_0', E_MCT1_ilacco1_0, 'exp'),
        ('ilacco1_1', E_MCT1_ilacco1_1, 'exp'),
        ('ilacco1_2', E_MCT1_ilacco1_2, 'exp')
    ],
    'BSG_MCT1': [
        ('ilacco1_0', E_BSG_MCT1_ilacco1_0, 'exp'),
        ('ilacco1_1', E_BSG_MCT1_ilacco1_1, 'exp'),
        ('ilacco1_2', E_BSG_MCT1_ilacco1_2, 'exp')
    ],
    'PfFNT': [
        ('ilacco1_0', E_PfFNT_ilacco1_0, 'exp'),
        ('ilacco1_1', E_PfFNT_ilacco1_1, 'exp'),
        ('ilacco1_2', E_PfFNT_ilacco1_2, 'exp')
    ],
    'PfFNT_FP': [
        ('ilacco1_0', E_PfFNT_FP_ilacco1_0, 'linear')
    ]
}

# Erstelle Subplots für jede Probe
num_probes = len(probe_data)
fig, axes = plt.subplots(1, num_probes, figsize=(20, 6))

# Sammle alle Transportraten für uniforme y-Achsenskalierung
all_transport_rates = []

# Erste Schleife: Sammle alle Werte für Skalierung
for probe_name, conditions in probe_data.items():
    for ilacco_name, df, fit_type in conditions:
        times = df["Time"].values
        means = df["Mittelwerte"].values
        
        mask = ~(np.isnan(times) | np.isnan(means))
        times_clean = times[mask]
        means_clean = means[mask]
        
        try:
            if fit_type == 'linear':
                coeffs = np.polyfit(times_clean, means_clean, 1)
                m_fit = coeffs[0]
                all_transport_rates.append(abs(m_fit))
            else:
                A_guess = means_clean[0] - means_clean[-1]
                C_guess = means_clean[-1]
                k_guess = 0.01
                
                popt, pcov = curve_fit(exp_decay, times_clean, means_clean, 
                                      p0=[A_guess, k_guess, C_guess],
                                      maxfev=10000)
                
                k_fit = popt[1]
                all_transport_rates.append(k_fit)
        except:
            pass

# Berechne einheitliche y-Achsen-Grenzen
if all_transport_rates:
    y_min = min(all_transport_rates)
    y_max = max(all_transport_rates)
    y_margin = (y_max - y_min) * 0.15
    y_lim = (y_min - y_margin, y_max + y_margin)
else:
    y_lim = None

# Zweite Schleife: Erstelle die Plots mit einheitlicher Skalierung
for idx, (probe_name, conditions) in enumerate(probe_data.items()):
    ax = axes[idx]
    
    ilacco_labels = []
    transport_vals = []
    transport_errs = []
    
    for ilacco_name, df, fit_type in conditions:
        times = df["Time"].values
        means = df["Mittelwerte"].values
        
        # Entferne NaN-Werte
        mask = ~(np.isnan(times) | np.isnan(means))
        times_clean = times[mask]
        means_clean = means[mask]
        
        try:
            if fit_type == 'linear':
                # Linearer Fit
                coeffs = np.polyfit(times_clean, means_clean, 1)
                m_fit = coeffs[0]
                
                # Berechne Unsicherheiten
                y_pred = np.polyval(coeffs, times_clean)
                residuals = means_clean - y_pred
                s_err = np.sqrt(np.sum(residuals**2) / (len(times_clean) - 2))
                t_mean = np.mean(times_clean)
                s_xx = np.sum((times_clean - t_mean)**2)
                m_err = s_err / np.sqrt(s_xx)
                
                ilacco_labels.append(ilacco_name)
                transport_vals.append(abs(m_fit))
                transport_errs.append(m_err)
            else:
                # Exponentieller Fit
                A_guess = means_clean[0] - means_clean[-1]
                C_guess = means_clean[-1]
                k_guess = 0.01
                
                popt, pcov = curve_fit(exp_decay, times_clean, means_clean, 
                                      p0=[A_guess, k_guess, C_guess],
                                      maxfev=10000)
                
                k_fit = popt[1]
                k_err = np.sqrt(pcov[1,1])
                
                ilacco_labels.append(ilacco_name)
                transport_vals.append(k_fit)
                transport_errs.append(k_err)
            
        except Exception as e:
            print(f'Fit für {probe_name} - {ilacco_name} fehlgeschlagen: {e}')
    
    # Plotte Histogramm für diese Probe
    x_pos = np.arange(len(ilacco_labels))
    bars = ax.bar(x_pos, transport_vals, yerr=transport_errs, capsize=5,
                  color='coral', edgecolor='black', linewidth=1.5, alpha=0.7)
    
    ax.set_title(probe_name, fontsize=12, fontweight='bold')
    ax.set_xlabel('ilacco-Variante', fontsize=11)
    if idx == 0:
        ax.set_ylabel('Transportrate [1/s]', fontsize=11)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(ilacco_labels, rotation=45, ha='right', fontsize=9)
    ax.grid(axis='y', alpha=0.3)
    
    # Setze einheitliche y-Achsenskalierung
    if y_lim:
        ax.set_ylim(y_lim)
    
    # Füge Werte über den Balken hinzu
    for i, (bar, val, err) in enumerate(zip(bars, transport_vals, transport_errs)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + err,
                f'{val:.5f}', ha='center', va='bottom', fontsize=8)

plt.suptitle('Vergleich der Transportraten über ilacco-Varianten für jede Probe', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()

outpath = os.path.join('figueres', 'Efflux_probe_across_ilacco_comparison.png')
plt.savefig(outpath, dpi=300, bbox_inches='tight')
print(f'Saved probe across ilacco comparison: {outpath}')
plt.show()
```
