```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
file1 = "1 mM Lactat Influx(2min)_MCT1_PfFNT_Zellen1_1_verdünnt.csv"
file2 = "1 mM Lactat Efflux(2min)_MCT1_PfFNT_Zellen1_1_verdünnt.csv"
df1 = pd.read_csv(file1, sep=';', decimal=',', skiprows=36, encoding='latin1')
df2 = pd.read_csv(file2, sep=';', decimal=',', skiprows=36, encoding='latin1')
# Ersetze "OVER" durch NaN in der gesamten DataFrame
df1 = df1.replace('OVER', np.nan)
df2 = df2.replace('OVER', np.nan)
df1.head()
```

```python
Kontrolle_I = dataframe1(df1, 3, 5, 6, 7)
MCT1_I= dataframe1(df1, 3, 8, 9, 10)
PfFNT_I = dataframe1(df1, 3, 11, 12, 13)

Kontrolle_E = dataframe1(df2, 3, 5, 6, 7)
MCT1_E= dataframe1(df2, 3, 8, 9, 10)
PfFNT_E = dataframe1(df2, 3, 11, 12, 13)
print(Kontrolle_I.head())
print(Kontrolle_E.head())
```

```python
plt.figure(figsize=(10, 6))

# Kontrolle Lactat
plt.errorbar(Kontrolle_I["Time"], Kontrolle_I["Mittelwerte"], 
             yerr=Kontrolle_I["Standardabweichung"],
             label="Kontrolle_I", linewidth=2, capsize=5, capthick=2)

# Kontrolle Hepes
plt.errorbar(MCT1_I["Time"], MCT1_I["Mittelwerte"],
             yerr=MCT1_I["Standardabweichung"],
             label="MCT1_I", linewidth=2, capsize=5, capthick=2)

# MCT1
plt.errorbar(PfFNT_I["Time"], PfFNT_I["Mittelwerte"],
             yerr=PfFNT_I["Standardabweichung"],
             label="PfFNT_I", linewidth=2, capsize=5, capthick=2)

plt.xlabel("Zeit (min)")
plt.ylabel("Efflux (dF/d0)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
